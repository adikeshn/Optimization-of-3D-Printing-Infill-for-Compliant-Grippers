```python
import cadquery as cq
from jupyter_cadquery import show

part = cq.importers.importStep("STEP_files/GripperForOpt_v2.step")


# Parameters
spacing = 5.0
rod_diameter = 1.0
width = 70  # X span
depth = 70   # Z span
thickness = 20  # extrusion along Y

grid = cq.Workplane("XZ")  # 👉 Build on XZ plane

# Vertical bars 
x = -width/2
while x <= width/2:
    grid = grid.union(
        cq.Workplane("XZ").center(x, 25)
        .rect(rod_diameter, depth)  # thin in X, tall in Z
        .extrude(-thickness)         # extrude along Y
    )
    x += spacing

# Horizontal bars 
z = -depth/2
while z <= depth/2:
    grid = grid.union(
        cq.Workplane("XZ").center(0, z+25)
        .rect(width, rod_diameter)  
        .extrude(-thickness)         
    )
    z += spacing

grid = grid.rotate((0, 0, 0), (0, 1, 0), 45)
# --- Trim the grid to fit inside the solid ---
infill_inside = grid.intersect(part)

# Option 1: show just the infill
show(infill_inside)

# Option 2: overlay (solid + infill together)
result = part.union(infill_inside)
show(result)
```
